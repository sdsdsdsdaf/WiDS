```python
from optuna.storages import JournalStorage
from optuna.storages.journal import JournalFileBackend

storage = JournalStorage(
    JournalFileBackend("gbsa_journal.log")
)

study = optuna.load_study(study_name="gbsa_survival", storage=storage)
```

```python
df = study.trials_dataframe()
df.head()
```

```python
df_sort = df.sort_values("value", ascending=False)
df_top20 = df_sort.head(20).drop(columns=["datetime_start", "datetime_complete", "duration", "state", "number",])
df_top20.to_csv("gbsa_top20.csv", index=False)
df_top20.head()
```
